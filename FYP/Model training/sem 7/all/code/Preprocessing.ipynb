```python
df =pd.read_csv("2_dataset.csv")
df
```

```python
x= df["ap_hi"]
y= df["ap_lo"]


# syshypo=int(115)
# sysnormal=int(125)

# diahypo=int(75)
# dianormal=int(85)



for i in df.index:
    if ((int (x[i])<int(115)) & (int (y[i])< int(80))):
        df["diagnosis"][i]= 1
        
    elif ((int (x[i])> int(130)) & (int (y[i])> int(80))) :
        df["diagnosis"][i]= 3
    else:
        df["diagnosis"][i]= 2  
        
df.value_counts("diagnosis")      
        
df.to_csv('2_dataset.csv', index=False)
```
